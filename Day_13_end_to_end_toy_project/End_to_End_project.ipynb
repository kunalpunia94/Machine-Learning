```python
#step-1->Preprocessing-remove unwanted column
df = df.iloc[:,1:]
df
```

```python
import matplotlib.pyplot as plt
```

```python
#EDA
plt.scatter(df['cgpa'],df['iq'],c=df['placement']) #yellow got placed
```

```python
#Extract input and output cols
#cgpa and iq are independent and placement is dependent it depends on cgpa and iq
x = df.iloc[:,0:2]  #gives all row and 0th and 1st column
y = df.iloc[:,-1]   #gives all row and last column
```

```python
x
```

```python
y
```

```python
#Train test split
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.1)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
#Scale the values between 1 and 0
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_train = scaler.fit_transform(x_train)
```

```python
x_train
```

```python
x_test = scaler.transform(x_test)
```

```python
x_test
```

```python
#Train the model
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
#now clf obejct have all the power of logistic regression
```

```python
#fit() function is used to train the data
clf.fit(x_train,y_train)
```

```python
#Evaluate the model
y_pred = clf.predict(x_test)
y_pred #this is predicted data
```

```python
y_test #this is test data
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
#how to plot decision boundary
#gwe will use library "mlxtend"
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x_train, y_train.values, clf=clf, legend=2)
```
