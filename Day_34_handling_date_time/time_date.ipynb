# working with dates

```python
#converting into datetime datatype
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

#### 1.Extract year

```python
date['date_year'] = date['date'].dt.year  #dt->datetime and year
```

```python
date.head()
```

#### 2.Extract month

```python
date['date_month_no'] = date['date'].dt.month  #dt->datetime and month
```

```python
date.head()
```

#### 3.extract month name

```python
date['date_month_name'] = date['date'].dt.month_name()  #dt->datetime and month_name
```

```python
date.head()
```

#### 4.Extract days

```python
date['date_day'] = date['date'].dt.day
```

```python
date.head()
```

#### 5.day of week

```python
date['date_dow'] = date['date'].dt.dayofweek
```

```python
date.head()
```

#### 6.day of week name

```python
date['date_dow_name'] = date['date'].dt.day_name()
```

```python
date.head()
```

```python
# is weekend?

date['date_is_weekend'] = np.where(date['date_dow_name'].isin(['Sunday', 'Saturday']), 1,0)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 7.Extract week of year

```python
date['date_week'] =  date['date'].dt.isocalendar().week

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 8.Extract Quarter

```python
date['quarter'] = date['date'].dt.quarter

date.drop(columns=['product_id','city_id','orders']).head()
```

#### 9.Extract semester

```python
date['semester'] = np.where(date['quarter'].isin([1,2]), 1, 2)

date.drop(columns=['product_id','city_id','orders']).head()
```

#### Extract Time elapsed between dates

```python
import datetime

today = datetime.datetime.today()

today
```

```python
today - date['date']
```

```python
(today - date['date']).dt.days
```

```python
#months passed
# Calculate months passed
date['months_passed'] = ((today.year - date['date'].dt.year) * 12 +
                         (today.month - date['date'].dt.month))

date.drop(columns=['product_id', 'city_id', 'orders']).head()
```
