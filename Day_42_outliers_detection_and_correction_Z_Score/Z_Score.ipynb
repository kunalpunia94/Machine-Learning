# Capping

```python
upper_limit = df['cgpa'].mean() + 3*df['cgpa'].std()
lower_limit = df['cgpa'].mean() - 3*df['cgpa'].std()
```

```python
upper_limit
```

```python
lower_limit
```

```python
#np.where(x,y,z) x->condition  y->if condition true what to do   z->if condition false then what to do
```

```python
df['cgpa'] = np.where(
    df['cgpa']>upper_limit,  #agar cgpa upper limmit se hayda hai
    upper_limit,             #replace by upper limit
    np.where(
        df['cgpa']<lower_limit,    #agar cgpa lower limit se kam hai
        lower_limit,               #replace by lower limit
        df['cgpa']            #else keep cgpa only
    )
)
```
