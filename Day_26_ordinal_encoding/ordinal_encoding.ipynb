```python
#age is numerical
#gender is nomial so we need to apply one hot encoding which we did not study yet
#review and education are ordinal so we will apply 
#perchased is in target so we need to apply label encoding on that
#so we will remove age and gender as of now....



df = df.iloc[:,2:]
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2) #in x first two cols and in y last col
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']]) #we have given in increasing order
```

```python
oe.fit(X_train)  #train ke upar fit karte hai but train and test dono ke upar transform karte hai
```

```python
X_train = oe.transform(X_train)
X_test = oe.transform(X_test)
```
